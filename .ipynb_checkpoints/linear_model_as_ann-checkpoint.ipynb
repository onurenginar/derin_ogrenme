```python
#Laoding required libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import datetime
import matplotlib.pyplot as plt
```

```python
tf.__version__
```

```python
#Reading kaggle radiation dataset
url="https://raw.githubusercontent.com/onurenginar/derin_ogrenme/master/SolarPrediction.csv"
data_solar = pd.read_csv(url)
```

```python
#take a look at dataset
print(data_solar)
```

```python
data_solar.columns
```

```python
#defining time index and removing unnucessary columns
ind = pd.to_datetime([datetime.datetime.fromtimestamp(i) for i in data_solar.UNIXTime])
data_solar.index =  ind
data_solar = data_solar.drop(['UNIXTime', 'Data', 'Time','TimeSunRise', 'TimeSunSet'],1)
```

```python
#defining input and output se (lets limit ourselves to first 15000 rows)
input = data_solar.iloc[0:15000,1:]
output = data_solar.Radiation[0:15000]
```

```python
#train test split
X_train, X_test, y_train, y_test = 	train_test_split(input, output, test_size=0.20,shuffle=True,random_state=42)

x_batches=X_train
x_test=X_test
y_batches=np.array(y_train)
y_test=np.array(y_test)
```

```python
#now first linear model
#model fitting
lin = LinearRegression()
lin_fit = lin.fit(x_batches, y_batches)
lin_pred = lin_fit.predict(x_test)
#error metric as normalized mean squared (has various definition)
mse = mean_squared_error(y_test, lin_pred)
nrmse_lin = np.power(mse,0.5)/np.mean(lin_pred)
print(nrmse_lin)
#vizualization of last 100 rows
plt.figure(figsize=(10, 7))
plt.plot(lin_pred[-100:], label = "prediction")
plt.plot(y_test[-100:], label = "real values")
plt.title("pred vs real")
plt.legend()
plt.show()
```

```python
#Keras Sequential ile doğrusal regresyon olarak yapay sinir ağı

#modelin katmanlar halinde yazılışı
model_nn = tf.keras.Sequential(
    [
     layers.Dense(1, activation = "linear", name = "layer1"),
     layers.Dense(1, activation = "linear", name = "layer2"),
    ]
)


#optimizasyon algoritmasının belirlenmesi
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
#maliyet fonksiyonunun belirlenmesi
model_nn.compile(optimizer=opt,
                 loss="mse",
                 metrics=["mse"])
#model fit edilmesi
model_nn.fit(x_batches, y_batches, validation_split = 0.1, batch_size = 256, epochs = 200)
#model ile tahmin
pred = model_nn.predict(x_test).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
#error metric as normalized mean squared (has various definition)
mse = mean_squared_error(y_test, pred)
nrmse_nn = np.power(mse,0.5)/np.mean(pred)
print(nrmse_nn)
#vizualization of last 100 rows
plt.figure(figsize=(10, 7))
plt.plot(pred[-100:], label = "prediction")
plt.plot(y_test[-100:], label = "real values")
plt.title("pred vs real")
plt.legend()
plt.show()
```

```python
# lets take a look at how different these two models
plt.figure(figsize=(10, 7))
plt.plot(pred[-100:], label = "nn_tahmin")
plt.plot(lin_pred[-100:], label = "lin_tahmin")
plt.title("tahmin_nn vs tahmin_lin")
plt.legend()
plt.show()

#and answer is almost surely same models as we claimed!!
```

```python
#now lets check the difference of the models outputs in a relative manner
diff_ratio = np.abs(nrmse_lin-nrmse_nn)/nrmse_lin
print(diff_ratio)
```
